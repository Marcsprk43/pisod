```python
def fast_shift(img, dx, dy):
    img_shape = img.shape

    img_new = np.zeros(img_shape)

    img_new[dy:, dx:,:] = img[0:(img_shape[0]-dy), 0:(img_shape[1]-dx), :]
```

```python
fast_shift(img, 5, 5)
```

```python
%timeit fast_shift(img, 500,500)
```
